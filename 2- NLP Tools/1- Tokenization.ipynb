```python
import spacy
from nltk.tokenize import word_tokenize
```

```python
text = '''Climate change threatens people with food and water scarcity, 
increased flooding, extreme heat, more disease, and economic loss. Human migration and conflict can be a result. 
The World Health Organization (WHO) calls climate change the greatest threat to global health in the 21st century.
'''
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
doc =  nlp(text)

print(doc)
print(type(doc))
```

```python
for word in doc :
    print(word) 
    print(word.shape_)
    print(word.is_digit)
    print(word.is_stop)    
    print('-'*20)
```

```python
wtk = [word for word in doc]
print(wtk)
```

```python
wtk = [word for word in word_tokenize(text)]
print(wtk)
```
